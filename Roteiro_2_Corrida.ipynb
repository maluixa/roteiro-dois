```python
# Fase 2 - Etapa 1 - Método dos Mínimos Quadrados 

import numpy as np

X = np.array([15.25, 16.38, 16.97, 18.12, 19.01])  #tempo (s)
Y = np.array([120, 130, 140, 150, 160])  #distância (m)

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,5))
plt.scatter(X, Y, color='r')
plt.xlabel('Tempo (s)')
plt.ylabel('Distância (m)')
plt.title('Atleta com velocidade constante')
plt.show()

n = len(X)
XY = X * Y
X2 = X ** 2

soma_X = sum(X)
soma_Y = sum(Y)
soma_X_2 = soma_X ** 2
soma_X2 = sum(X2)
soma_XY = sum(XY)

c1 = (soma_X * soma_Y - n * soma_XY) / (soma_X_2 - n * soma_X2)
c2 = (soma_XY * soma_X - soma_X2 * soma_Y) / (soma_X_2 - n * soma_X2)

print("O valor do somatório soma_X é:", soma_X)
print("O valor do somatório soma_Y é:", soma_Y)
print("O valor do somatório soma_X_2 é:", soma_X_2)
print("O valor do somatório soma_X2 é:", soma_X2)
print("O valor do somatório soma_XY é:", soma_XY)

print("O valor da constante c1 (velocidade) é:", c1)
print("O valor da constante c2 (posição inicial) é:", c2)

print("A velocidade constante do corredor é:", c1, "m/s")

Ya = []
for i in range(n):
    y = c1 * X[i] + c2
    Ya.append(y)

fig = plt.figure(figsize=(10,5))
plt.scatter(X, Y, color='r')
plt.xlabel('Tempo (s)')
plt.ylabel('Distância (m)')
plt.title('Gráfico de ajuste - Atleta com velocidade constante')
plt.plot(X, Ya, linestyle='solid', color='blue')
plt.legend(['Reta ajustada', 'Medidas'])
plt.show()
```
