```python
import numpy as np

X = np.array([0.6, 1.2, 1.8, 2.4, 3.0, 3.6])
Y = np.array([5.90, 23.90, 53.50, 95.00, 148.50, 213.80])

import matplotlib.pyplot as plt
import math

fig = plt.figure(figsize=(10,5))
plt.scatter(X, Y, color='r')
plt.xlabel('Deformao (m)')
plt.ylabel('Energia Potencial (J)')
plt.title('Estudo do comportamento da mola do elevador')
plt.show()

ln_X = np.log(X)
ln_Y = np.log(Y)

plt.scatter(ln_X, ln_Y, c='r', alpha=0.6)
plt.title('log-log', fontsize=15)
plt.xlabel('log(Deformao)', fontsize=13)
plt.ylabel('log(Energia Potencial)', fontsize=13)
plt.grid(True, which='major', linewidth=2, linestyle='-')
plt.minorticks_on()
plt.grid(True, which='minor', linewidth=1, linestyle='--')

plt.show()

n = len(X)
ln_XY = ln_X * ln_Y
ln_X2 = ln_X ** 2

soma_X = sum(ln_X)
soma_Y = sum(ln_Y)
soma_X_2 = soma_X ** 2
soma_X2 = sum(ln_X2)
soma_XY = sum(ln_XY)

a1 = (soma_X * soma_Y - n * soma_XY) / (soma_X_2 - n * soma_X2)
a2 = (soma_XY * soma_X - soma_X2 * soma_Y) / (soma_X_2 - n * soma_X2)

print("O valor do somatrio soma_X :", soma_X)
print("O valor do somatrio soma_Y :", soma_Y)
print("O valor do somatrio soma_X_2 :", soma_X_2)
print("O valor do somatrio soma_X2 :", soma_X2)
print("O valor do somatrio soma_XY :", soma_XY)

print("O valor da constante a1 (potncia) :", a1)
print("O valor da constante a2 :", a2)

k = 2 * math.exp(a2)
print("O valor da constante da mola k :", k)


```

```python
T = np.array([0.3, 0.6, 0.9, 1.2, 1.5, 1.8, 2.1, 2.4, 2.7])
I = np.array([-15.69, -27.69, 37.42, -1.67, -36.11, 30.00, 12.57, -39.86, 18.70])

import math
f = 60
w = 2 * math.pi * f

xa = []
for i in T:
    fa = math.cos(w * i * math.pi / 180)  # transformar para radianos
    xa.append(fa)
xa = np.array(xa)

fig = plt.figure(figsize=(10,5))
plt.scatter(xa, I, color='r')
plt.xlabel('Tempo (s)')
plt.ylabel('Corrente (A)')
plt.title('Estudo da corrente alternada do circuito')
plt.show()

n = len(T)
xaI = xa * I
xa2 = xa ** 2

soma_xa = sum(xa)
soma_I = sum(I)
soma_xa_2 = soma_xa ** 2
soma_xa2 = sum(xa2)
soma_xaI = sum(xaI)

a1 = (soma_xa * soma_I - n * soma_xaI) / (soma_xa_2 - n * soma_xa2)
a2 = (soma_xaI * soma_xa - soma_xa2 * soma_I) / (soma_xa_2 - n * soma_xa2)

print("O valor do somatrio soma_xa :", soma_xa)
print("O valor do somatrio soma_I :", soma_I)
print("O valor do somatrio soma_xa_2 :", soma_xa_2)
print("O valor do somatrio soma_xa2 :", soma_xa2)
print("O valor do somatrio soma_xaI :", soma_xaI)

print("O valor da constante a1 :", a1)
print("O valor da constante a2 (amplitude mxima da corrente) :", a2)
```
